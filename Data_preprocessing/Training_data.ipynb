```python
import numpy as np
import pandas as pd
```

```python
input_data = pd.read_csv('X_Train_Data_Input.csv')
input_data.head(10)
```

```python
target = pd.read_csv('Y_Train_Data_Target.csv')
target.head(10)
```

```python
target['target'].value_counts()
```

```python
input_data.shape
```

```python
target.shape
```

```python
data = pd.merge(input_data, target, on = 'ID')
```

```python
data.head(10)
```

```python
data.shape
```

```python
# Drop the ID column:
data = data.drop(columns=['ID'])
data.shape
```

```python
data
```

```python
# Missing values in each of the columns
missing_value = data.isnull().sum()
missing_value
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = (missing_value / len(data)) * 100

missing_data = pd.DataFrame({
    'Column': data.columns,
    'Missing_Percentage': missing_percentage
})

# Filter out columns where the missing percentage is 0
missing_data_non_zero = missing_data[missing_data['Missing_Percentage'] > 0].reset_index(drop=True)

(missing_data_non_zero)
```

```python
# Drop the column 9 cauz 93 % of values are actually missing
data = data.drop(columns='Column9')
```

```python
# Mean imputation for numerical columns
data.fillna(data.mean(), inplace=True)
```

```python
data.head(10)
```

```python
data.to_csv('Training.csv', index=False)
```
